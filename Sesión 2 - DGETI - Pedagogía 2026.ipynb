## **Taller de valores**: Dilemas Éticos en Ciberseguridad y Robótica
---

### Definición de Dilemas Éticos
> Un dilema ético es una situación en la que una persona o una organización se enfrenta a una elección entre dos o más opciones, cada una de las cuales presenta un conflicto con principios morales o valores fundamentales. No existe una solución obvia o "correcta" que satisfaga todos los principios, y la decisión tomada a menudo implica sacrificar un valor en favor de otro.

### Dilemas Éticos en Ciberseguridad
> La ciberseguridad, al proteger sistemas y datos, a menudo se encuentra en la encrucijada de la seguridad, la privacidad, la libertad y la legalidad. Aquí un ejemplo clave:

*   **Hacking Ético vs. Privacidad y Legalidad:**

    *   **Contexto:**
    > Un experto en ciberseguridad (hacker ético o 'white hat') descubre una vulnerabilidad crítica en los sistemas de una empresa o gobierno que podría ser explotada por actores maliciosos. Para demostrar la gravedad de la falla y forzar su corrección, podría verse tentado a acceder a información sensible o a realizar una intrusión que, aunque bien intencionada, podría ser vista como una violación de la privacidad o incluso como una actividad ilegal si no cuenta con el consentimiento explícito y previo del propietario del sistema.

    *   **Dilema:**
    > * ¿Se debe priorizar la seguridad general (revelando la vulnerabilidad de forma contundente, quizás intrusiva) por encima de la privacidad de los datos o la estricta legalidad de los métodos de prueba?
    > * ¿Hasta qué punto es aceptable "romper" algo para demostrar que está roto, especialmente si eso implica acceder a datos de usuarios?

### Dilemas Éticos en Robótica
---
> La robótica, al introducir máquinas autónomas en la sociedad, plantea cuestiones fundamentales sobre el trabajo, la responsabilidad, la autonomía y el impacto social. Un ejemplo prominente es:

*   **Automatización y Empleo vs. Eficiencia y Progreso:**
    *   **Contexto:**
    > Una empresa decide implementar robots y sistemas de inteligencia artificial para automatizar una parte significativa de su línea de producción o de sus servicios (ej. chatbots para atención al cliente), lo que resulta en una mayor eficiencia, reducción de costos, y potencialmente, productos de mejor calidad o más accesibles.
    *   **Dilema:**
    > * ¿Cómo equilibrar los beneficios económicos y de eficiencia de la automatización con el impacto social de la pérdida de empleos a gran escala?
    > * ¿Tiene la empresa o la sociedad una responsabilidad moral de reentrenar a los trabajadores desplazados o de crear nuevas oportunidades?
    > * ¿Debería el progreso tecnológico frenarse si conlleva un costo humano significativo en términos de desempleo y desigualdad?
    > * Este dilema pone en tensión el valor del progreso económico y la innovación frente al bienestar social y la estabilidad laboral.

## Mapeo de conceptos y vinculación curricular (transversal y longitudinal)
---

La integración de los **Conceptos Centrales de la Educación para el Desarrollo Sostenible (COCEDS)** en el currículo es fundamental para formar profesionales conscientes y proactivos. Esta integración se puede realizar de manera **transversal** (aplicando los COCEDS a través de diferentes asignaturas en un mismo nivel educativo) y **longitudinal** (desarrollando la comprensión de los COCEDS a lo largo de los diferentes niveles y años de estudio).

Los **COCEDS** clave para esta vinculación son:

*   **Acción Transformadora:** Capacitar a los estudiantes para ser agentes de cambio, no solo entendiendo los problemas sino actuando para resolverlos.
*   **Acción Sistémica:** Fomentar una comprensión holística de cómo los diferentes elementos de un sistema (tecnológico, social, ambiental) interactúan entre sí.
*   **Acción Estratégica:** Desarrollar la capacidad de planificar y ejecutar acciones a largo plazo, considerando el impacto futuro de las decisiones actuales.
*   **Acción Global-Social:** Promover la conciencia sobre la interconexión global de los problemas y la responsabilidad social de las soluciones tecnológicas.

### Ejemplos

#### **Actividad X**: Auditoría de Ciberseguridad con Impacto Energético

*   **Especialidad:** Ciberseguridad y Gestión de Infraestructuras

*   **Descripción:** Los estudiantes realizarán una auditoría simulada o real de ciberseguridad en un pequeño centro de datos (o una infraestructura virtualizada). No solo identificarán vulnerabilidades, sino que también analizarán el impacto energético de las medidas de seguridad implementadas (ej. cifrado de datos, sistemas de detección de intrusiones, redundancia de sistemas) y propondrán optimizaciones que mejoren la seguridad sin disparar el consumo de energía.

*   **COCEDS Integrados:**

    *   **Acción Sistémica:** Comprender la interconexión entre seguridad, rendimiento y consumo energético.

    *   **Acción Estratégica:** Desarrollar un plan de mejora de seguridad que sea sostenible a largo plazo en términos de recursos energéticos.

    *   **Acción Transformadora:** Proponer soluciones innovadoras que optimicen ambos aspectos.

#### **Actividad Y**: Desarrollo de Políticas de Uso Ético y Sostenible de IA en Robots

*   **Especialidad:** Robótica, Inteligencia Artificial y Ética Tecnológica

*   **Descripción:** En equipos, los estudiantes diseñarán un conjunto de políticas de uso para un robot con IA en un escenario específico (ej. un robot de asistencia para personas mayores, un dron de reparto autónomo). Deberán considerar aspectos éticos (privacidad, autonomía, responsabilidad), sociales (impacto en el empleo, accesibilidad) y ambientales (consumo de energía del robot, ciclo de vida de los materiales). Presentarán sus políticas y justificarán sus decisiones basándose en los COCEDS.

*   **COCEDS Integrados:**

    *   **Acción Global-Social:** Considerar el impacto de la IA y la robótica en la sociedad y el planeta.

    *   **Acción Estratégica:** Formular políticas que guíen el desarrollo y uso de la tecnología de manera responsable a largo plazo.

    *   **Acción Sistémica:** Analizar cómo las decisiones de diseño y uso afectan múltiples dimensiones (ética, social, ambiental).

#### **Actividad Z**: Hackathon de Soluciones Verdes para Infraestructuras Digitales

*   **Especialidad:** Ciberseguridad, Redes y Sostenibilidad Digital

*   **Descripción:** Un hackathon donde los estudiantes trabajen en equipos para idear y prototipar soluciones tecnológicas que reduzcan la huella de carbono de las infraestructuras digitales. Los proyectos pueden incluir desde el desarrollo de algoritmos de enrutamiento más eficientes, herramientas de monitoreo de consumo energético para servidores, hasta diseños de software que minimicen el uso de recursos computacionales. Se evaluará la viabilidad técnica, el impacto ambiental y la alineación con los principios de desarrollo sostenible.

*   **COCEDS Integrados:**

    *   **Acción Transformadora:** Generar soluciones tangibles para desafíos ambientales en el ámbito digital.

    *   **Acción Sistémica:** Entender cómo las decisiones de diseño de software y hardware afectan el consumo energético global.

    *   **Acción Global-Social:** Contribuir a la reducción de la huella de carbono digital, un problema de alcance global.
