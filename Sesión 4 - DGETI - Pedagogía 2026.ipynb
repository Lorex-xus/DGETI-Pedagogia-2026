## Aprendizaje activo en métodos de caso

### Estructura y delimitaciones

#### Componentes Clave y Delimitaciones del Método de Casos
El método de casos es una herramienta pedagógica poderosa que fomenta el aprendizaje activo y el desarrollo de habilidades críticas. Para su implementación efectiva, es fundamental comprender sus componentes esenciales, fases, roles, objetivos y limitaciones.

#### Componentes Esenciales de un Caso de Estudio
Un caso de estudio efectivo se compone de varios elementos fundamentales que guían el proceso de aprendizaje:
*   **Documento del Caso:** Es el corazón del método. Presenta una situación o problema real o hipotético, con datos relevantes, descripciones de personajes, contexto y, a menudo, dilemas o decisiones a tomar. Debe ser lo suficientemente complejo como para permitir múltiples interpretaciones y soluciones, y lo suficientemente claro para que los participantes puedan entender el escenario. Puede incluir textos, gráficos, tablas, y materiales multimedia.
*   **Análisis Individual:** Cada participante revisa el documento del caso de forma independiente, identificando los problemas principales, analizando la información disponible, desarrollando posibles soluciones y preparando argumentos para defender su postura.
*   **Discusión Grupal (Pequeños Grupos):** Los participantes se reúnen en grupos pequeños para compartir sus análisis individuales, debatir diferentes perspectivas, refinar sus argumentos y llegar a un consenso o a una serie de recomendaciones. Esta fase es crucial para el desarrollo de habilidades de comunicación y trabajo en equipo.
*   **Discusión Plenaria (Gran Grupo):** Todos los grupos se unen para presentar sus análisis y soluciones al resto de la clase, moderados por el facilitador. Aquí se exponen y comparan las distintas aproximaciones, se profundiza en la comprensión del caso y se exploran las complejidades del problema.
*   **Conclusiones y Debriefing (Cierre):** El facilitador sintetiza los aprendizajes clave, aclara conceptos, y conecta las discusiones del caso con la teoría relevante. Esta fase es vital para consolidar el conocimiento y reflexionar sobre el proceso de toma de decisiones y las habilidades desarrolladas.

#### Delimitaciones y Limitaciones del Método de Casos
Aunque es muy beneficioso, el método de casos presenta ciertas limitaciones:
*   **Tiempo Requerido:** Es un método intensivo en tiempo, tanto para la preparación del caso por el facilitador como para el análisis y discusión por parte de los participantes. No es adecuado para cubrir rápidamente grandes volúmenes de contenido.
*   **Dependencia de la Habilidad del Facilitador:** El éxito del método depende en gran medida de la capacidad del docente para guiar la discusión, hacer las preguntas adecuadas y manejar las dinámicas de grupo. Un facilitador inexperto puede disminuir la efectividad.
*   **Posible Subjetividad en la Interpretación:** Dado que los casos suelen ser ambiguos y abiertos a múltiples interpretaciones, puede haber una alta subjetividad en el análisis, lo que podría generar frustración si los participantes buscan una única respuesta "correcta".
*   **Falta de una Única Respuesta Correcta:** A diferencia de otros métodos pedagógicos, el método de casos rara vez tiene una solución única y definitiva, lo que puede ser desafiante para estudiantes acostumbrados a problemas con soluciones claras.
*   **Transferibilidad de las Soluciones al Mundo Real:** Aunque se basa en situaciones reales, la implementación de las soluciones desarrolladas en el aula a contextos reales puede variar y no siempre ser directamente aplicable debido a factores externos no contemplados en el caso.

## Método de Casos en la Práctica

El método de casos, en la práctica, se desarrolla a través de una serie de fases interconectadas y asigna roles específicos para maximizar el aprendizaje, orientándose hacia objetivos pedagógicos claros.

### Fases Típicas del Método de Casos
El método de casos se estructura generalmente en las siguientes fases:
1.  **Preparación:** Introducción al método, distribución del caso y tiempo para la lectura y análisis individual por parte de los participantes. El facilitador establece los objetivos de aprendizaje.
2.  **Análisis Individual:** Cada participante lee y analiza el caso por su cuenta, identificando problemas, generando hipótesis y desarrollando una estrategia inicial.
3.  **Discusión en Grupos Pequeños:** Los participantes se reúnen en equipos para compartir y debatir sus análisis individuales, buscando construir una comprensión colectiva y propuestas de solución.
4.  **Discusión Plenaria:** Los grupos presentan sus hallazgos y recomendaciones a la clase, se discuten las diferencias y similitudes, y se explora la gama de posibles soluciones.
5.  **Cierre y Debriefing:** El facilitador guía una reflexión final sobre el proceso, los resultados, las lecciones aprendidas y su aplicabilidad a otros contextos.

### Roles Clave
*   **Estudiantes/Participantes:** Su rol es activo y central. Deben leer, analizar, investigar, debatir, tomar decisiones, comunicarse eficazmente y participar activamente en todas las fases. Son responsables de su propio aprendizaje y del aprendizaje colectivo.
*   **Facilitador (Docente):** Actúa como guía y moderador, no como dador de respuestas. Su función es estimular el debate, hacer preguntas desafiantes, gestionar el tiempo, asegurar la participación equitativa, mantener el enfoque en los objetivos de aprendizaje y resumir las principales conclusiones. Debe ser experto en el contenido y en la dinámica de grupos.

### Objetivos de Aprendizaje Principales
La aplicación del método de casos en tecnología educativa busca alcanzar diversos objetivos, incluyendo:
*   **Desarrollo del Pensamiento Crítico:** Capacidad para analizar situaciones complejas, identificar problemas subyacentes, evaluar información y formar juicios razonados.
*   **Habilidades de Resolución de Problemas:** Destreza para abordar desafíos de manera estructurada, generar soluciones creativas y seleccionar la más adecuada.
*   **Toma de Decisiones:** Habilidad para evaluar opciones bajo incertidumbre y justificar las elecciones realizadas.
*   **Comunicación Efectiva:** Mejora en la capacidad de expresar ideas claramente, escuchar activamente, persuadir y negociar en un entorno grupal.
*   **Aplicación de Conocimientos Teóricos a Situaciones Prácticas:** Puente entre la teoría aprendida y su implementación en escenarios del mundo real, especialmente en el contexto de la tecnología educativa.
*   **Trabajo en Equipo y Liderazgo:** Fomento de la colaboración, la gestión de conflictos y el desarrollo de habilidades de liderazgo dentro de un equipo.

### Ejemplo de Caso: Robótica/Automatización

### Título del Caso:
**Fallo Crítico en el Brazo Robótico de una Línea de Ensamblaje Automotriz**

### Contexto y Descripción del Problema:
La empresa "Autoconsumo SA" opera una avanzada línea de ensamblaje para vehículos eléctricos. Recientemente, el brazo robótico principal (modelo "RoboArm X-Series") en la estación de soldadura crítica ha comenzado a presentar fallos intermitentes de posicionamiento, resultando en soldaduras defectuosas en la carrocería de los vehículos. Estos fallos han provocado una alta tasa de rechazo en el control de calidad final, interrupciones en la producción y un aumento significativo en los costos de retrabajo. La gerencia está alarmada, ya que la inactividad de esta estación afecta la producción de toda la planta.

### Datos Relevantes:
*   **Registros de Fallos**: Historial de los últimos 3 meses del RoboArm X-Series, incluyendo códigos de error, fechas y horas de los fallos de posicionamiento, y duración de las interrupciones.
*   **Manual Técnico del RoboArm X-Series**: Especificaciones de software (firmware v3.2), hardware (servomotores, sensores de posición) y procedimientos de calibración.
*   **Diagramas Eléctricos y Neumáticos**: Esquemas de conexión de los componentes del brazo robótico.
*   **Registros de Mantenimiento Preventivo**: Fechas de las últimas calibraciones, inspecciones y reemplazos de componentes.
*   **Producción Diaria**: Datos de la cantidad de unidades producidas y rechazadas antes y después del inicio de los fallos.
*   **Video del Proceso**: Grabaciones de la estación de soldadura durante operaciones normales y durante un fallo intermitente de posicionamiento.
*   **Entrevistas**: Resúmenes de conversaciones con operarios y el técnico de mantenimiento principal, quienes han reportado ruidos inusuales o movimientos erráticos antes de los fallos.

### Desafíos y Preguntas Orientadoras:
*   **Identificación de Causa Raíz**: ¿Cuál es la causa o las causas fundamentales del fallo intermitente de posicionamiento en el RoboArm X-Series?
*   **Impacto Operacional**: ¿Cómo se cuantifica el impacto económico y de producción de estos fallos?
*   **Opciones de Solución**: ¿Qué soluciones técnicas se pueden implementar a corto y largo plazo para corregir el problema (ej., ajuste de software, reemplazo de hardware, recalibración, mejora del mantenimiento)?
*   **Evaluación de Riesgos**: ¿Cuáles son los riesgos asociados a cada solución propuesta (costo, tiempo de inactividad, complejidad, seguridad)?
*   **Prevención Futura**: ¿Qué medidas se pueden tomar para prevenir la recurrencia de este tipo de fallos en el futuro?

### Tareas para el Estudiante:
1.  **Análisis de Datos**: Examinar los datos proporcionados para identificar patrones, correlaciones y anomalías relacionadas con los fallos.
2.  **Diagnóstico del Problema**: Formular hipótesis sobre las posibles causas raíz y justificarlas con la evidencia disponible.
3.  **Propuesta de Solución**: Desarrollar al menos dos soluciones técnicas viables para corregir el fallo, detallando los pasos de implementación y los recursos necesarios.
4.  **Análisis de Costo-Beneficio**: Evaluar las soluciones propuestas en términos de costo, impacto en la producción, durabilidad y retorno de inversión.
5.  **Plan de Acción**: Crear un plan de acción recomendado para la gerencia de Autoconsumo SA, incluyendo la solución elegida, un cronograma estimado y las justificaciones técnicas y económicas.

### Ejemplo de Caso: Ciberseguridad/Implementación

### Título del Caso:
**Brecha de Datos en Plataforma Educativa Online: Respuesta y Fortalecimiento**

### Contexto y Descripción del Problema:
"EduTech Global", una plataforma educativa online con millones de usuarios a nivel mundial, ha sufrido recientemente una brecha de datos significativa. Un atacante logró explotar una vulnerabilidad en un módulo de autenticación de terceros, obteniendo acceso a datos sensibles de estudiantes y docentes, incluyendo nombres, correos electrónicos, historiales de cursos y, en algunos casos, datos de facturación cifrados. El incidente fue detectado varias horas después de haber ocurrido. La empresa enfrenta ahora no solo un daño reputacional masivo y la posible fuga de usuarios, sino también multas regulatorias bajo GDPR y CCPA, y la urgente necesidad de restaurar la confianza de sus usuarios y asegurar su plataforma para evitar futuros ataques.

### Datos Relevantes:
*   **Registros de Servidor Web (Apache/Nginx)**: Logs de acceso y errores del período del incidente.
*   **Logs del Sistema de Detección de Intrusiones (IDS)**: Alertas generadas en las horas previas y durante la brecha.
*   **Configuraciones de Firewall**: Reglas de entrada y salida del perímetro de red y de la aplicación.
*   **Reporte Preliminar de Forense Digital**: Detalles de la vulnerabilidad explotada y los vectores de ataque iniciales.
*   **Política de Privacidad y Términos de Servicio de EduTech Global**: Documentos legales que definen el manejo de datos de usuario.
*   **Contrato con el Proveedor del Módulo de Autenticación de Terceros**: Acuerdos de nivel de servicio (SLA) y responsabilidades.
*   **Estructura del Equipo de Seguridad**: Organigrama y roles del equipo de respuesta a incidentes de EduTech Global.
*   **Comunicaciones Internas**: Correos electrónicos del equipo de gestión de crisis tras la detección del incidente.

### Desafíos y Preguntas Orientadoras:
*   **Respuesta a Incidentes**: ¿Cómo debe priorizarse la respuesta inicial para contener y erradicar la amenaza, minimizando el daño continuo?
*   **Análisis Forense**: ¿Qué pasos adicionales se necesitan para comprender completamente el alcance de la brecha y asegurar que todos los puntos de entrada hayan sido identificados y cerrados?
*   **Obligaciones Legales y de Cumplimiento**: ¿Qué pasos debe tomar EduTech Global para cumplir con las regulaciones de protección de datos (GDPR, CCPA) en relación con la notificación y mitigación?
*   **Comunicación Estratégica**: ¿Cómo debe comunicarse la brecha a los usuarios, la prensa y las autoridades reguladoras para mitigar el daño reputacional?
*   **Fortalecimiento de la Seguridad**: ¿Qué medidas de seguridad a largo plazo (técnicas, procesos, políticas) deben implementarse para prevenir incidentes similares en el futuro?
*   **Plan de Recuperación**: ¿Qué pasos son necesarios para restaurar completamente la confianza de los usuarios y asegurar la continuidad del negocio?

### Tareas para el Estudiante:
1.  **Plan de Respuesta a Incidentes**: Desarrollar un plan de acción paso a paso para la contención, erradicación y recuperación del incidente.
2.  **Análisis de Vulnerabilidades**: Identificar posibles vulnerabilidades adicionales en la infraestructura de EduTech Global basándose en la información disponible y proponer soluciones preventivas.
3.  **Estrategia de Comunicación**: Diseñar un borrador de comunicado para los usuarios afectados y un plan de comunicación para los medios.
4.  **Recomendaciones de Seguridad**: Proponer un conjunto de mejoras de seguridad (tecnológicas, políticas, formación) para fortalecer la postura de ciberseguridad de EduTech Global a mediano y largo plazo.
5.  **Evaluación de Impacto Regulatorio**: Analizar las implicaciones de la brecha bajo GDPR y CCPA y sugerir acciones para asegurar el cumplimiento.

## Ejemplos con código

```python
import pandas as pd
import matplotlib.pyplot as plt


# Porcentaje de acceso a electricidad por año*ciudad*escuela
df = pd.read_csv("https://raw.githubusercontent.com/Lorex-xus/DGETI-Pedagogia-2026/refs/heads/main/Files/school_electricity_access.csv")
df.head(10)
```

```python
# Ciuidades ordenadas alfabeticamente con porcentaje de acceso promedio en los años registrados
df[["Country_Name", "Electricity_Access_Percent"]].groupby("Country_Name").mean().sort_values(by="Electricity_Access_Percent", ascending=False)
```

### Tendencia Global de Acceso a Electricidad

Calcular el porcentaje promedio de acceso a la electricidad por año y visualizar la tendencia global a lo largo del tiempo con un gráfico de líneas para entender cómo ha evolucionado el acceso a la electricidad en las escuelas globalmente.


```python
promedios = df.groupby('Year')['Electricity_Access_Percent'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(promedios['Year'], promedios['Electricity_Access_Percent'], marker='o', linestyle='-')
plt.title('Tendencia Global del Porcentaje Promedio de Acceso a Electricidad en Escuelas por Año')
plt.xlabel('Año')
plt.ylabel('Porcentaje Promedio de Acceso a Electricidad')
plt.grid(True)
plt.xticks(promedios['Year'].astype(int))
plt.tight_layout()
plt.show()
```

### Top 10 Países con Mayor Acceso a Electricidad

Mostrar los 10 países con el mayor porcentaje de acceso a la electricidad para ese año. Visualizar esto con un gráfico de barras horizontales, incluyendo el nombre del país y el porcentaje de acceso.


```python
# Considerando que solo hay 3 datos del  último año, será tomado el 2019 como referencia

ultimo_anno = 2019 #@param {type:"slider", min:2000, max:2020, step:1}
df_anno = df[df['Year'] == ultimo_anno] # Discriminación con mascara booleana
top_10 = df_anno.sort_values(by='Electricity_Access_Percent', ascending=False).head(10)

plt.figure(figsize=(12, 7))
plt.barh(top_10['Country_Name'], top_10['Electricity_Access_Percent'], color='skyblue')
plt.xlabel('Porcentaje de Acceso a Electricidad')
plt.ylabel('País')
plt.title(f'Top 10 Países con Mayor Acceso a Electricidad en Escuelas en {ultimo_anno}')
plt.gca().invert_yaxis() # Valor más alto arriba
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```

### Top 10 Países con Menor Acceso a Electricidad

Mostrar los 10 países con el menor porcentaje de acceso a la electricidad para ese año (excluyendo países con 0% si son muchos). Visualizar esto con un gráfico de barras horizontales, incluyendo el nombre del país y el porcentaje de acceso.


```python

ultimo_anno = 2019 #@param {type:"slider", min:2000, max:2020, step:1}
df_anno = df[df['Year'] == ultimo_anno] # Discriminación con mascara booleana
df_ultimo_anno = df_anno[df_anno['Electricity_Access_Percent'] > 0]
bottom_10_countries = df_ultimo_anno.sort_values(by='Electricity_Access_Percent', ascending=True).head(10)

plt.figure(figsize=(12, 7))
plt.barh(bottom_10_countries['Country_Name'], bottom_10_countries['Electricity_Access_Percent'], color='lightcoral')
plt.xlabel('Porcentaje de Acceso a Electricidad')
plt.ylabel('País')
plt.title(f'Top 10 Países con Menor Acceso a Electricidad en Escuelas en {ultimo_anno}')
plt.gca().invert_yaxis() # Valor menor hasta arriba
plt.xlim(0, 100)
plt.tight_layout()
plt.show()
```
